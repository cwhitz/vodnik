```python
import requests
import json
```

## Test Next Line Generation

```python
t = """Mr. Hops went to the market. As usual, he was looking for his favorite orange treat: carrots. As he went along the stream,
he saw something unusual."""

next_line_payload = {
    "text": t,
    "additional_instructions": "Something violent and bloody happens.",
    "word_count": 50
}

response = requests.post("http://127.0.0.1:8000" + "/generate/next", json=next_line_payload)

print("Next Line Generation Response:")
if response.status_code != 200:
    print("Error:", response.status_code, response.text)
else:
    print(response.json()['generated_text'])
```
